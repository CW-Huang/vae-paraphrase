```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
colours = [
    (1, 0, 0),
    (0, 1, 0),
    (0, 0, 1),
    (0, 1, 1)
]
import math
def plot_attention(ax, words, attention_matrix):
    for k in xrange(attention_matrix.shape[0]):
        for i in xrange(attention_matrix.shape[1]):
            for j in xrange(attention_matrix.shape[2]):
                ax.plot(
                    [i, j], [0, 1],
                    linewidth=5,
                    color=colours[k] + (attention_matrix[k, j, i],)
                )
    ax.set_xticks(range(len(words)))
    ax.set_xticklabels(words)
    ax.yaxis.set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.xaxis.set_tick_params(labeltop='on')
    for label in ax.xaxis.get_ticklabels():
        label.set_rotation(90)
        label.set_fontsize(15)
        
```

```python
import model
import data_io
from theano_toolkit.parameters import Parameters
import tracker
import theano.tensor as T
import theano
idx2word, word2idx = data_io.load_dictionary('dict.pkl')
unk_idx = len(word2idx)

embedding_count = len(word2idx) + 2
embedding_size = 256

P = Parameters()
    
P.embedding = np.random.randn(embedding_count, embedding_size)
_, encoder = model.build_encoder(
    P,
    embedding_size=embedding_size,
    annotation_size=256,
    latent_size=128
)
X_2 = T.imatrix('X_2')
encoder(P.embedding[X_2.T], T.neq(X_2.T, -1))
attn_0 = tracker.get_variable(P, 'trans_0_attn_attention')
attn_1 = tracker.get_variable(P, 'trans_1_attn_attention')
attn_2 = tracker.get_variable(P, 'trans_2_attn_attention')
attn_3 = tracker.get_variable(P, 'trans_3_attn_attention')

compute_attention = theano.function(
    inputs=[X_2],
    outputs=[attn_0, attn_1, attn_2, attn_3]
)
```

```python
P.load('model.pkl', strict=False)
line = "the police shot pablo escobar on tuesday ." 
tokens = line.split()
token_ids = np.array([[word2idx.get(w, unk_idx)
                       for w in tokens]], dtype=np.int32)
val_attn = compute_attention(token_ids)
```

```python
plt.figure(figsize=(20, 5))
ax = plt.axes()
attn = val_attn[1][0, :2]
plot_attention(ax, tokens, attn)
```
